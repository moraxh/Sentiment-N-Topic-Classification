```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

def plotSentiments(filename):
    df = pd.read_csv(filename)
    sentiments = df["Sentiment"]
    plt.hist(sentiments)
    plt.xlabel(filename)
    plt.show()
```

```python

plotSentiments("classified/sentiments/1erTrimestre_Claudia.csv")
plotSentiments("classified/sentiments/1erTrimestre_Maynez.csv")
plotSentiments("classified/sentiments/1erTrimestre_Xochitl.csv")

plotSentiments("classified/sentiments/2doTrimestre_Claudia.csv")
plotSentiments("classified/sentiments/2doTrimestre_Maynez.csv")
plotSentiments("classified/sentiments/2doTrimestre_Xochitl.csv")
```
